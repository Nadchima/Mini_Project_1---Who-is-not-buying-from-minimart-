# **Who is not buying from minimart?**
Observe the following data: https://github.com/mafudge/datasets/tree/master/minimart

You are to display names of customers who did not buy from minimart in any given month.`

HINT

Access the raw data from this base url: https://raw.githubusercontent.com/mafudge/datasets/refs/heads/master/minimart/

```python
import pandas as pd
customers = pd.read_csv('https://raw.githubusercontent.com/mafudge/datasets/refs/heads/master/minimart/customers.csv')
customers
```

**รวมตาราง customers และ purchases ในแต่ละเดือน**

ตัวอย่าง: รวมตาราง customers และ purchases_jan โดยใช้ customer_id เป็นกุญแจ (key)

1. รวมตาราง customers และ purchases_jan โดยใช้ customer_id เป็นกุญแจ (key)
2. การตรวจสอบคอลัมน์ order_id ว่าเป็นค่า ว่าง (NaN) หรือไม่
*   ถ้ามีค่า → แสดงว่าลูกค้าคนนั้นมีการสั่งซื้อ
*   ถ้าเป็น NaN → แสดงว่าลูกค้าคนนั้นไม่ได้สั่งซื้อในเดือนมกราคม

3. สำหรับลูกค้าที่ไม่มีการสั่งซื้อ (order_id เป็น NaN) ให้เพิ่มคอลัมน์ใหม่ชื่อ no_order_month
แล้วใส่ค่า 'January' เพื่อบอกว่าไม่ได้สั่งซื้อในเดือนนี้


```python
purchases_jan = pd.read_csv('https://raw.githubusercontent.com/mafudge/datasets/refs/heads/master/minimart/purchases-jan.csv')
purchases_jan
```

```python
purchases_feb = pd.read_csv('https://raw.githubusercontent.com/mafudge/datasets/refs/heads/master/minimart/purchases-feb.csv')
purchases_feb
```

```python
purchases_mar = pd.read_csv('https://raw.githubusercontent.com/mafudge/datasets/refs/heads/master/minimart/purchases-mar.csv')
purchases_mar
```

```python
purchases_apr = pd.read_csv('https://raw.githubusercontent.com/mafudge/datasets/refs/heads/master/minimart/purchases-apr.csv')
purchases_apr
```

```python
# January
customer01 = pd.merge(customers, purchases_jan, left_on='customer_id', right_on='customer_id', how='left')
order_null = customer01['order_id'].isna()
customer01.loc[order_null, 'no_order_month'] = 'January'
customer01
```

```python
# February
customer02 = pd.merge(customers, purchases_feb, left_on='customer_id', right_on='customer_id', how='left')
order_null = customer02['order_id'].isna()
customer02.loc[order_null, 'no_order_month'] = 'February'
customer02
```

```python
# March
customer03 = pd.merge(customers, purchases_mar, left_on='customer_id', right_on='customer_id', how='left')
order_null = customer03['order_id'].isna()
customer03.loc[order_null, 'no_order_month'] = 'March'
customer03
```

```python
# April
customer04 = pd.merge(customers, purchases_apr, left_on='customer_id', right_on='customer_id', how='left')
order_null = customer04['order_id'].isna()
customer04.loc[order_null, 'no_order_month'] = 'April'
customer04
```

```python
total_customer = pd.concat([customer01, customer02, customer03, customer04], ignore_index=True)
total_customer.sort_values('customer_id')
```

```python
# นำค่าที่มี NaN ออกในแถว no_order_month
total_customer = total_customer.dropna(subset=['no_order_month'])
total_customer
```

```python
total_customer = total_customer.drop(columns=['order_id','order_date','order_amount'])
```

```python
total_customer.sort_values('customer_id')
```

```python
from os import name

from IPython.display import display
from ipywidgets import interact_manual

@ interact_manual(name='', month=["January", "February", "March", "April", "All"])
def on_click(month, name):
  display(month)
  display(name)

  if month == "All":                                                                  # เงื่อนไขที่เลือกเดือนทั้งหมด
    if name == "":
      filtered_customers = total_customer                                             # แสดงผลการค้นหาของทุกเดือนและทุกคนที่ไม่ได้มาซื้อ
    else:
      filtered_customers = total_customer[total_customer['firstname'] == name]        # แสดงผลการค้นหาชื่อคนที่ไม่ได้มาซื้อในทุกเดือน
  else:                                                                               # เงื่อนไขที่ไม่ได้เลือกเดือนทั้งหมด แต่เลือกตามเดือนนั้นๆ
    if name == "":
      filtered_customers = total_customer[total_customer['no_order_month'] == month]  # เลือกเดือนที่ลูกค้าไม่ได้มาซื้อเดือนนั้นๆ โดยไม่ระบุชื่อลูกค้าในการค้นหา
    else:
      filtered_customers = total_customer[(total_customer['no_order_month'] == month) & (total_customer['firstname'] == name)]    # แสดงผลจากการเลือกเดือนและชื่อลูกค้าที่ไม่ได้มาซื้อ


  if not filtered_customers.empty:
    display(filtered_customers)
  else:
    display("No customers found for the selected month and name.")
```
